```python
import pandas as pd
import numpy as np
```

Data regarding the capacity of all 7 crematoriums in Wuhan city per day.

```python
crematorium_data={'crematorium_name':['Hankou Crematorium','Wuchang Crematorium','Qingshan Crematorium','Huangpi Crematorium','Jiangxia Crematorium','Caidian Crematorium','Xinzhou Crematorium'],
                  'hearses_in_service':[20,24,5,14,0,15,3],
                  'cremator':[30,15,8,7,7,12,5],
                 'capatity_per_day':[720,360,192,168,168,288,120],
                 'urns_per_day':[500,500,500,500,500,500,500]}
cd_df=pd.DataFrame(crematorium_data)
cd_df
```

```python
cd_df['capatity_per_day'].sum()
```

Strick city-wise lock-down started from 01/23/2020, till 03/23/2020 family can start picking up the ashes from crematoriums and offered "free urns" from the Authority. Each crematorium had to give out 500 urns each day
 till 04/03/2020 before Chinese traditional "tomb-sweeping" day which is to commemorate the dead. 
As the hearse picked up my father's body, I talked to one of the staff, I was told all the crematoriums were operating 24/7 for days, and the bodies were still piling up. I started to calculate the real numbers within the first 14 days of lock-down, which was the worst two weeks as I felt and experienced there, also due to time intervals between infection and death of a given patient, I used 14 days as my reference.

```python
# create column 'bodies_burned_14days'
cd_df['bodies_burned_14days']=cd_df['capatity_per_day']*14
```

```python
cd_df['bodies_burned_14days'].sum()
```

```python
# create column 'urns_giveout_12days'
cd_df['urns_giveout_12days']=cd_df['urns_per_day']*12
cd_df['urns_giveout_12days'].sum()
```

The actual death number would range from 28000 to 42000 during the 2 months of lockdown, which matches the number that I was told by the crematorium staff on the day I picked up my father's ashes, he said roughly 20000-30000 bodies were cremated by his fellow workers.
Also, the death rate at the very beginning of the pandemic, released by the National Health Commission of PRC, local CDC, and Caixin Insight, collected and published by Caixin Data, was 5%.

```python
death_rate={'city_name':['Hubei(Province)','Wuhan','Qianjiang',
                        'Huanggang',
                        'Ezhou',
                        'Jingmen',
                        'Yichang',
                        'Xiantao',
                        'Huangshi',
                        'Xiaogan',
                        'Suizhou',
                        'Xiangyang',
                        'Jingzhou',
                        'Tianmen',
                        'Enshi',
                        'Xianning',
                        'Shiyan'],
           'death_rate':[0.047,0.05,0.045,0.043,0.042,0.042,0.039,0.038,0.037,0.036,0.034,0.033,0.032,0.03,0.028,0.018,0.012]}
dr_df=pd.DataFrame(death_rate)
dr_df
```

```python
import matplotlib.pyplot as plt
import numpy as np

city_name=dr_df['city_name'].unique()
colors=['red' if i =='Wuhan' else 'blue' for i in city_name]
dr_df.plot.bar(x='city_name',y='death_rate',rot=90,color=colors,legend=None)
plt.ylabel('Death Rate')
plt.xlabel('City Names')
plt.title('Death Rate of the Cities in Hubei Province')
plt.show()
    
```

As the graph shows above, Wuhan had the highest death rate due to the dense population and the most important position of central transportation. According to the death number I came out above, which ranges from 28000 to 42000, the infection toll would be:

```python
death_toll=cd_df['bodies_burned_14days'].sum()

infetion_toll=death_toll/0.05
infetion_toll
```
